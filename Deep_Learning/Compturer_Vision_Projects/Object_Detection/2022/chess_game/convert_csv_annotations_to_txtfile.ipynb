```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/drive/MyDrive/Tech_Neuron/Machine_Learning/Deep_Learning/Computer_Vision/Object_detection/Chess_detection/train.csv")
df.head()
```

```python
df_new = pd.DataFrame()
df_new['class'] = df['class'].copy()
df_new['xmin'] = df['xmin'].copy().astype('float')
df_new['ymin'] = df['ymin'].copy().astype('float')
df_new['xmax'] = df['xmax'].copy().astype('float')
df_new['ymax'] = df['ymax'].copy().astype('float')
df_new.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df_new['class'] = le.fit_transform(df_new['class'])
df_new.head(5)
```

```python
norm_list = []
for i in range(3742):
  l = list(df_new.iloc[i])[1:]
  norm = [round(float(j)/sum(l),6) for j in l]
  norm_list.append(norm)
  df_new['xmin'][i] = norm[0]
  df_new['ymin'][i] = norm[1]
  df_new['xmax'][i] = norm[2]
  df_new['ymax'][i] = norm[3]

df_new.head()
```

```python
my_list = list(df_new.iloc[0])[1:]
mystring = str(df_new['class'][0])+" "+str(my_list[0])+" "+str(my_list[1])+" "+str(my_list[2])+" "+str(my_list[3])
mystring
```
